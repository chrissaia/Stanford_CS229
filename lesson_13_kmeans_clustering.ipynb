# Lesson 13: K-means Clustering## Objectives- Implement k-means from scratch.- Visualize cluster assignments.- Track distortion (inertia) over iterations and across k.

## From the notesK-means alternates between assigning points to the nearest centroid and updating centroids as means of assigned points.

## IntuitionK-means optimizes within-cluster squared distance by coordinate descent on assignments and centroids.

## DataWe generate three Gaussian blobs.

```python
import numpy as npimport matplotlib.pyplot as pltnp.random.seed(42)
```

```python
# Synthetic blobscenters = [(-3,0), (0,3), (3,0)]X = np.vstack([np.random.multivariate_normal(c, np.eye(2)*0.5, size=60) for c in centers])
```

## Implementation: k-means

```python
def kmeans(X, k, num_iters=20):    idx = np.random.choice(len(X), k, replace=False)    centroids = X[idx]    history = []    for _ in range(num_iters):        dists = np.linalg.norm(X[:, None, :] - centroids[None, :, :], axis=2)        labels = np.argmin(dists, axis=1)        new_centroids = np.vstack([X[labels == j].mean(axis=0) for j in range(k)])        centroids = new_centroids        distortion = np.mean(np.min(dists, axis=1))        history.append(distortion)    return labels, centroids, np.array(history)
```

## Experiments

```python
labels, centroids, history = kmeans(X, k=3)
```

## Visualizations

```python
plt.figure(figsize=(6,4))for k in range(3):    plt.scatter(X[labels==k,0], X[labels==k,1], alpha=0.7, label=f"cluster {k}")plt.scatter(centroids[:,0], centroids[:,1], color="black", marker="x", s=100, label="centroids")plt.xlabel("x1")plt.ylabel("x2")plt.title("K-means clusters")plt.legend()plt.show()plt.figure(figsize=(6,4))plt.plot(history)plt.xlabel("iteration")plt.ylabel("distortion")plt.title("K-means convergence")plt.show()
```

## Takeaways- K-means is simple but sensitive to initialization.- Distortion decreases monotonically with iterations.
