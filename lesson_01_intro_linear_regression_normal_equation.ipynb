## From the notes: notation + objectiveWe use the CS229 notation: training data \(\{(x^{(i)}, y^{(i)})\}_{i=1}^m\), feature dimension \(n\), parameters \(	heta\), and hypothesis \(h_	heta(x) = 	heta^T x\) with \(x_0 = 1\).Least-squares objective:\[J(	heta) = rac{1}{2m} \sum_{i=1}^m (h_	heta(x^{(i)}) - y^{(i)})^2.\]Normal equation solution:\[	heta = (X^T X)^{-1} X^T y.\]

## IntuitionLinear regression finds the line (or hyperplane) that minimizes squared error. The normal equation gives the closed-form minimizer when \(X^T X\) is invertible.
