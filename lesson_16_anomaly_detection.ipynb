# Lesson 16: Anomaly Detection## Objectives- Fit a Gaussian model for normal data.- Score points by probability density.- Visualize anomalies and thresholding.

## From the notesAnomaly detection estimates \(p(x)\) on normal data and flags low-probability points.

## IntuitionRare points under the learned density are treated as anomalies.

## DataWe generate normal data plus a few outliers.

```python
import numpy as npimport matplotlib.pyplot as pltnp.random.seed(42)
```

```python
# Normal data + anomaliesnormal = np.random.multivariate_normal([0,0], np.eye(2), size=150)anom = np.random.multivariate_normal([5,5], np.eye(2)*0.4, size=6)X = np.vstack([normal, anom])
```

## Implementation: Gaussian anomaly detection

```python
mu = normal.mean(axis=0)Sigma = np.cov(normal.T)Sigma_inv = np.linalg.pinv(Sigma)def gaussian_density(X, mu, Sigma_inv):    diff = X - mu    exp_term = np.exp(-0.5 * np.sum(diff @ Sigma_inv * diff, axis=1))    denom = np.sqrt((2*np.pi)**X.shape[1] * (np.linalg.det(np.linalg.pinv(Sigma_inv)) + 1e-9))    return exp_term / denomscores = gaussian_density(X, mu, Sigma_inv)threshold = np.percentile(scores, 5)flags = scores < threshold
```

## Visualizations

```python
plt.figure(figsize=(6,4))plt.scatter(X[:,0], X[:,1], c=flags, cmap="coolwarm", alpha=0.7)plt.xlabel("x1")plt.ylabel("x2")plt.title("Anomaly flags")plt.show()plt.figure(figsize=(6,4))plt.hist(scores, bins=20)plt.axvline(threshold, color="red", label="threshold")plt.xlabel("p(x)")plt.ylabel("count")plt.title("Density scores")plt.legend()plt.show()
```

## Takeaways- Density-based detection requires a good model of normal behavior.- Threshold selection controls precision/recall of anomalies.
